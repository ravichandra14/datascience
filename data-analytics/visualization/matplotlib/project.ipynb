```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv(
    "../../../datasets/cleaned/cleaned_crop_production.csv"
)

# ==========================================
# 1. LINE PLOT
# ==========================================

yearly_production = df.groupby(
    "crop_year"
)["production"].sum()

plt.figure(figsize=(10, 5))

plt.plot(
    yearly_production.index,
    yearly_production.values,
    marker="o"
)

plt.title("Year Wise Production")
plt.xlabel("Year")
plt.ylabel("Production")

plt.show()
```

```python
# ==========================================
# 2. BAR CHART
# ==========================================

top_crops = df.groupby(
    "crop"
)["production"].sum().sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(12, 6))

plt.bar(
    top_crops.index,
    top_crops.values
)

plt.title("Top 10 Crops Production")
plt.xlabel("Crop")
plt.ylabel("Production")

plt.xticks(rotation=45)

plt.show()

# ==========================================
# 3. HORIZONTAL BAR CHART
# ==========================================

plt.figure(figsize=(10, 6))

plt.barh(
    top_crops.index,
    top_crops.values
)

plt.title("Horizontal Bar Chart")

plt.show()

# ==========================================
# 4. HISTOGRAM
# ==========================================

plt.figure(figsize=(10, 5))

plt.hist(
    df["production"],
    bins=30
)

plt.title("Production Distribution")
plt.xlabel("Production")
plt.ylabel("Frequency")

plt.show()

# ==========================================
# 5. SCATTER PLOT
# ==========================================

plt.figure(figsize=(10, 5))

plt.scatter(
    df["area"],
    df["production"]
)

plt.title("Area vs Production")
plt.xlabel("Area")
plt.ylabel("Production")

plt.show()

# ==========================================
# 6. PIE CHART
# ==========================================

top5 = top_crops.head(5)

plt.figure(figsize=(8, 8))

plt.pie(
    top5.values,
    labels=top5.index,
    autopct="%1.1f%%"
)

plt.title("Top 5 Crop Share")

plt.show()

# ==========================================
# 7. BOX PLOT
# ==========================================

plt.figure(figsize=(8, 5))

plt.boxplot(df["production"])

plt.title("Production Box Plot")

plt.show()

# ==========================================
# 8. AREA PLOT
# ==========================================

plt.figure(figsize=(10, 5))

plt.fill_between(
    yearly_production.index,
    yearly_production.values
)

plt.title("Area Plot")
plt.xlabel("Year")
plt.ylabel("Production")

plt.show()

# ==========================================
# 9. STEM PLOT
# ==========================================

sample = yearly_production.head(10)

plt.figure(figsize=(10, 5))

plt.stem(
    sample.index,
    sample.values
)

plt.title("Stem Plot")

plt.show()

# ==========================================
# 10. STACK PLOT
# ==========================================

years = [2018, 2019, 2020, 2021]

rice = [100, 120, 130, 140]
wheat = [90, 100, 110, 120]

plt.figure(figsize=(10, 5))

plt.stackplot(
    years,
    rice,
    wheat,
    labels=["Rice", "Wheat"]
)

plt.legend()

plt.title("Stack Plot")

plt.show()

# ==========================================
# 11. STEP PLOT
# ==========================================

plt.figure(figsize=(10, 5))

plt.step(
    yearly_production.index,
    yearly_production.values
)

plt.title("Step Plot")

plt.show()

# ==========================================
# 12. ERROR BAR PLOT
# ==========================================

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
error = [1, 2, 1, 3]

plt.figure(figsize=(8, 5))

plt.errorbar(
    x,
    y,
    yerr=error,
    fmt="o"
)

plt.title("Error Bar Plot")

plt.show()

# ==========================================
# 13. SUBPLOTS
# ==========================================

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 0].set_title("Line")

ax[0, 1].bar(["A", "B"], [10, 20])
ax[0, 1].set_title("Bar")

ax[1, 0].hist(df["production"], bins=20)
ax[1, 0].set_title("Histogram")

ax[1, 1].scatter(df["area"], df["production"])
ax[1, 1].set_title("Scatter")

plt.tight_layout()

plt.show()

# ==========================================
# 14. SAVE FIGURE
# ==========================================

plt.figure(figsize=(10, 5))

plt.plot(
    yearly_production.index,
    yearly_production.values
)

plt.title("Saved Plot")

plt.savefig("production_plot.png")

plt.show()

# ==========================================
# 15. STYLE CUSTOMIZATION
# ==========================================

plt.figure(figsize=(10, 5))

plt.plot(
    yearly_production.index,
    yearly_production.values,
    linestyle="--",
    marker="o",
    linewidth=2
)

plt.grid(True)

plt.title("Customized Plot")

plt.show()
```
