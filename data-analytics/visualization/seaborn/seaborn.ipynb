```python
# ==========================================
# COMPLETE SEABORN + MATPLOTLIB VISUALIZATION
# CROP PRODUCTION ANALYSIS PROJECT
# ==========================================

# ------------------------------------------
# JUPYTER NOTEBOOK FIX
# ------------------------------------------



# ------------------------------------------
# IMPORT LIBRARIES
# ------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------------------
# LOAD DATASET
# ------------------------------------------

df = pd.read_csv(
    "../../../datasets/cleaned/cleaned_crop_production.csv"
)

# ------------------------------------------
# BASIC CHECKS
# ------------------------------------------

print("FIRST 5 ROWS")
print(df.head())

print("\nDATASET SHAPE")
print(df.shape)

print("\nCOLUMN NAMES")
print(df.columns)

# ------------------------------------------
# SEABORN STYLE
# ------------------------------------------

sns.set_style("whitegrid")
```

```python
# ==========================================
# 1. LINE PLOT
# ==========================================

yearly_production = df.groupby(
    "crop_year"
)["production"].sum().reset_index()

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=yearly_production,
    x="crop_year",
    y="production",
    marker="o"
)

plt.title(
    "Year Wise Crop Production",
    fontsize=18
)

plt.xlabel("Year", fontsize=14)
plt.ylabel("Production", fontsize=14)

plt.show()
```

```python
# ==========================================
# 2. BAR PLOT
# ==========================================

top_crops = df.groupby(
    "crop"
)["production"].sum().sort_values(
    ascending=False
).head(10).reset_index()

plt.figure(figsize=(14, 6))

sns.barplot(
    data=top_crops,
    x="crop",
    y="production"
)

plt.title(
    "Top 10 Crop Production",
    fontsize=18
)

plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# 3. HISTOGRAM
# ==========================================

plt.figure(figsize=(12, 6))

sns.histplot(
    df["production"],
    bins=30,
    kde=True
)

plt.title(
    "Production Distribution",
    fontsize=18
)

plt.xlabel("Production")
plt.ylabel("Frequency")

plt.show()

# ==========================================
# 4. SCATTER PLOT
# ==========================================

sample_df = df.sample(1000)

plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=sample_df,
    x="area",
    y="production"
)

plt.title(
    "Area vs Production",
    fontsize=18
)

plt.xlabel("Area")
plt.ylabel("Production")

plt.show()
```

```python
# ==========================================
# 5. BOX PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.boxplot(
    x=df["production"]
)

plt.title(
    "Production Box Plot",
    fontsize=18
)

plt.show()

# ==========================================
# 6. HEATMAP
# ==========================================

numeric_df = df.select_dtypes(
    include=["int64", "float64"]
)

correlation = numeric_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)

plt.title(
    "Correlation Heatmap",
    fontsize=18
)

plt.show()
```

```python
# ==========================================
# 7. COUNT PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.countplot(
    data=df,
    x="season"
)

plt.title(
    "Season Distribution",
    fontsize=18
)

plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================
# 8. PIE CHART
# ==========================================

top5 = top_crops.head(5)

plt.figure(figsize=(8, 8))

plt.pie(
    top5["production"],
    labels=top5["crop"],
    autopct="%1.1f%%"
)

plt.title(
    "Top 5 Crop Share",
    fontsize=18
)

plt.show()
```

```python
# ==========================================
# 9. VIOLIN PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.violinplot(
    x=df["production"]
)

plt.title(
    "Production Violin Plot",
    fontsize=18
)

plt.show()
```

```python
# ==========================================
# 10. REGRESSION PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.regplot(
    data=sample_df,
    x="area",
    y="production"
)

plt.title(
    "Regression Plot",
    fontsize=18
)

plt.show()
```

```python
# ==========================================
# 11. KDE PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.kdeplot(
    df["production"],
    fill=True
)

plt.title(
    "Production Density Plot",
    fontsize=18
)

plt.show()
```

```python
# ==========================================
# 12. PAIR PLOT
# ==========================================

pair_df = sample_df[[
    "area",
    "production",
    "crop_year"
]]

sns.pairplot(pair_df)

plt.show()
```

```python
# ==========================================
# 13. SAVE GRAPH
# ==========================================

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=yearly_production,
    x="crop_year",
    y="production"
)

plt.title("Saved Production Plot")

plt.savefig(
    "crop_production_plot.png")

plt.show()

# ==========================================
# PROJECT COMPLETED
# ==========================================

print("\nALL VISUALIZATIONS GENERATED SUCCESSFULLY")
```
