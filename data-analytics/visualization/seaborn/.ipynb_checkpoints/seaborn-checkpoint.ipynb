```python
# ==========================================
# SEABORN VISUALIZATION PRACTICE
# ==========================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------------------------
# LOAD DATASET
# ------------------------------------------

df = pd.read_csv(
    "../../../datasets/cleaned/cleaned_crop_production.csv"
)

# Seaborn Style
sns.set_style("whitegrid")

# ==========================================
# 1. LINE PLOT
# ==========================================

yearly_production = df.groupby(
    "crop_year"
)["production"].sum().reset_index()

plt.figure(figsize=(10, 5))

sns.lineplot(
    x="crop_year",
    y="production",
    data=yearly_production
)

plt.title("Year Wise Production")

plt.show()

# ==========================================
# 2. BAR PLOT
# ==========================================

top_crops = df.groupby(
    "crop"
)["production"].sum().sort_values(
    ascending=False
).head(10).reset_index()

plt.figure(figsize=(12, 6))

sns.barplot(
    x="crop",
    y="production",
    data=top_crops
)

plt.xticks(rotation=45)

plt.title("Top 10 Crops Production")

plt.show()

# ==========================================
# 3. HISTOGRAM
# ==========================================

plt.figure(figsize=(10, 5))

sns.histplot(
    df["production"],
    bins=30,
    kde=True
)

plt.title("Production Distribution")

plt.show()

# ==========================================
# 4. SCATTER PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.scatterplot(
    x="area",
    y="production",
    data=df
)

plt.title("Area vs Production")

plt.show()

# ==========================================
# 5. BOX PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.boxplot(
    x=df["production"]
)

plt.title("Production Box Plot")

plt.show()

# ==========================================
# 6. VIOLIN PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.violinplot(
    x=df["production"]
)

plt.title("Production Violin Plot")

plt.show()

# ==========================================
# 7. COUNT PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.countplot(
    x="season",
    data=df
)

plt.xticks(rotation=45)

plt.title("Season Count")

plt.show()

# ==========================================
# 8. HEATMAP
# ==========================================

numeric_df = df.select_dtypes(
    include=["int64", "float64"]
)

correlation = numeric_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    correlation,
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()

# ==========================================
# 9. PAIR PLOT
# ==========================================

sample_df = df[[
    "area",
    "production",
    "crop_year"
]].sample(500)

sns.pairplot(sample_df)

plt.show()

# ==========================================
# 10. KDE PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.kdeplot(
    df["production"],
    fill=True
)

plt.title("Production Density")

plt.show()

# ==========================================
# 11. JOINT PLOT
# ==========================================

sns.jointplot(
    x="area",
    y="production",
    data=df,
    kind="scatter"
)

plt.show()

# ==========================================
# 12. REGRESSION PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.regplot(
    x="area",
    y="production",
    data=df
)

plt.title("Regression Plot")

plt.show()

# ==========================================
# 13. STRIP PLOT
# ==========================================

sample = df.sample(1000)

plt.figure(figsize=(12, 6))

sns.stripplot(
    x="season",
    y="production",
    data=sample
)

plt.xticks(rotation=45)

plt.title("Strip Plot")

plt.show()

# ==========================================
# 14. SWARM PLOT
# ==========================================

plt.figure(figsize=(12, 6))

sns.swarmplot(
    x="season",
    y="production",
    data=sample
)

plt.xticks(rotation=45)

plt.title("Swarm Plot")

plt.show()

# ==========================================
# 15. CAT PLOT
# ==========================================

sns.catplot(
    x="season",
    y="production",
    data=sample,
    kind="bar",
    height=6,
    aspect=2
)

plt.xticks(rotation=45)

plt.show()

# ==========================================
# 16. FACET GRID
# ==========================================

g = sns.FacetGrid(
    sample,
    col="season"
)

g.map(
    sns.histplot,
    "production"
)

plt.show()

# ==========================================
# 17. DISPLOT
# ==========================================

sns.displot(
    df["production"],
    kde=True
)

plt.show()

# ==========================================
# 18. ECDF PLOT
# ==========================================

plt.figure(figsize=(10, 5))

sns.ecdfplot(
    data=df,
    x="production"
)

plt.title("ECDF Plot")

plt.show()

# ==========================================
# 19. RELPLOT
# ==========================================

sns.relplot(
    x="area",
    y="production",
    data=sample
)

plt.show()

# ==========================================
# 20. LM PLOT
# ==========================================

sns.lmplot(
    x="area",
    y="production",
    data=sample
)

plt.show()

# ==========================================
# IMPORTANT SEABORN PLOTS
# ==========================================

"""
1. lineplot()
2. barplot()
3. histplot()
4. scatterplot()
5. boxplot()
6. heatmap()
7. pairplot()
8. regplot()
9. countplot()
10. violinplot()
"""

# ==========================================
# END
# ==========================================
```
