```python
# ==========================================
# CROP PRODUCTION DATA CLEANING USING PANDAS
# ==========================================

import pandas as pd
import numpy as np

# ------------------------------------------
# 1. LOAD DATASET
# ------------------------------------------

df = pd.read_csv("../../../datasets/raw/crop_production.csv")

# ------------------------------------------
# 2. BASIC INFORMATION
# ------------------------------------------

print("\nFIRST 5 ROWS")
print(df.head())

print("\nDATASET INFO")
print(df.info())

print("\nSHAPE")
print(df.shape)

print("\nCOLUMN NAMES")
print(df.columns)

# ------------------------------------------
# 3. CLEAN COLUMN NAMES
# ------------------------------------------

df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

print("\nCLEANED COLUMN NAMES")
print(df.columns)

# ------------------------------------------
# 4. CHECK MISSING VALUES
# ------------------------------------------

print("\nMISSING VALUES")
print(df.isnull().sum())

# ------------------------------------------
# 5. REMOVE DUPLICATES
# ------------------------------------------

print("\nDUPLICATES BEFORE")
print(df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("\nDUPLICATES AFTER")
print(df.duplicated().sum())

# ------------------------------------------
# 6. CLEAN STRING COLUMNS
# ------------------------------------------

string_columns = [
    "state_name",
    "district_name",
    "season",
    "crop"
]

for col in string_columns:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.strip()
    df[col] = df[col].str.lower()

print("\nSTRING CLEANING COMPLETED")

# ------------------------------------------
# 7. HANDLE MISSING VALUES
# ------------------------------------------

df["production"] = pd.to_numeric(
    df["production"],
    errors="coerce"
)

median_production = df["production"].median()

df["production"].fillna(
    median_production)

df["area"] = pd.to_numeric(
    df["area"],
    errors="coerce"
)

median_area = df["area"].median()

df["area"].fillna(
    median_area
)

print("\nMISSING VALUES AFTER CLEANING")
print(df.isnull().sum())

# ------------------------------------------
# 8. REMOVE INVALID VALUES
# ------------------------------------------

df = df[df["area"] > 0]
df = df[df["production"] > 0]

print("\nAFTER REMOVING INVALID VALUES")
print(df.shape)

# ------------------------------------------
# 9. CONVERT YEAR COLUMN
# ------------------------------------------

df["crop_year"] = pd.to_numeric(
    df["crop_year"],
    errors="coerce"
)

# ------------------------------------------
# 10. OUTLIER DETECTION
# ------------------------------------------

Q1 = df["production"].quantile(0.25)
Q3 = df["production"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df = df[
    (df["production"] >= lower_limit) &
    (df["production"] <= upper_limit)
]

print("\nAFTER OUTLIER REMOVAL")
print(df.shape)

# ------------------------------------------
# 11. RESET INDEX
# ------------------------------------------

df.reset_index(drop=True, inplace=True)

# ------------------------------------------
# 12. FINAL DATASET
# ------------------------------------------

print("\nFINAL DATASET")
print(df.head())

print("\nFINAL SHAPE")
print(df.shape)

# ------------------------------------------
# 13. SAVE CLEANED DATASET
# ------------------------------------------

df.to_csv(
    "../../../datasets/cleaned/cleaned_crop_production.csv",
    index=False
)

print("\nCLEANED DATASET SAVED SUCCESSFULLY")

# ==========================================
# END
# ==========================================
```
