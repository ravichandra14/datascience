```python
# ==========================================
# CUSTOMER SEGMENTATION SYSTEM
# ==========================================

# ------------------------------------------
# IMPORT LIBRARIES
# ------------------------------------------

import os
import warnings

warnings.filterwarnings("ignore")

os.environ["OMP_NUM_THREADS"] = "1"

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ==========================================
# CREATE FOLDERS
# ==========================================

os.makedirs(
    "../../../datasets/cleaned/customer_segmentation",
    exist_ok=True
)

os.makedirs(
    "../../../datasets/processed/customer_segmentation",
    exist_ok=True
)

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv(
    "../../../datasets/raw/mall_customers/Mall_Customers.csv"
)

# ==========================================
# CLEANING
# ==========================================

df = df.drop_duplicates()

df = df.dropna()

# ==========================================
# SAVE CLEANED DATASET
# ==========================================

df.to_csv(
    "../../../datasets/cleaned/customer_segmentation/cleaned_customers.csv",
    index=False
)

# ==========================================
# FEATURE SELECTION
# ==========================================

X = df[[
    "Annual Income (k$)",
    "Spending Score (1-100)"
]]

# ==========================================
# SCALING
# ==========================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# ==========================================
# SAVE PROCESSED DATA
# ==========================================

pd.DataFrame(X_scaled).to_csv(
    "../../../datasets/processed/customer_segmentation/X_scaled.csv",
    index=False
)

# ==========================================
# ELBOW METHOD
# ==========================================

wcss = []

for i in range(1, 11):

    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(X_scaled)

    wcss.append(model.inertia_)

# ==========================================
# ELBOW PLOT
# ==========================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker="o"
)

plt.title("Elbow Method")

plt.xlabel("Clusters")
plt.ylabel("WCSS")

plt.show()

# ==========================================
# FINAL MODEL
# ==========================================

model = KMeans(
    n_clusters=5,
    random_state=42
)

clusters = model.fit_predict(
    X_scaled
)

# ==========================================
# ADD CLUSTERS
# ==========================================

df["Cluster"] = clusters

# ==========================================
# SAVE RESULTS
# ==========================================

df.to_csv(
    "../../../datasets/processed/customer_segmentation/clustered_data.csv",
    index=False
)

# ==========================================
# VISUALIZATION
# ==========================================

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df["Annual Income (k$)"],
    y=df["Spending Score (1-100)"],
    hue=df["Cluster"],
    palette="deep"
)

plt.title("Customer Segmentation")

plt.show()

# ==========================================
# END
# ==========================================
```
